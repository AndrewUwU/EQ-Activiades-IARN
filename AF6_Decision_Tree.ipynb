```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, classification_report
```

```python
# Cargar el dataset
data = load_breast_cancer()
X = data.data
y = data.target

print(f"Dataset cargado: {data.target_names}")
```

```python
# División de datos (80% entrenamiento, 20% prueba)
train_data, test_data, train_labels, test_label = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print(f"Tamaño del conjunto de entrenamiento: {len(train_data)}")
print(f"Tamaño del conjunto de prueba: {len(test_data)}")
```

```python
# Crear el modelo
modelo = DecisionTreeClassifier(random_state=42)

# Entrenar el modelo
modelo.fit(train_data, train_labels)
print("Modelo entrenado:D")
```

```python
# Realizar predicciones
predicciones = modelo.predict(test_data)
```

```python
# Obtener matriz de confusión
matriz = confusion_matrix(test_label, predicciones)
disp = ConfusionMatrixDisplay(confusion_matrix=matriz, display_labels=data.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión - Clasificación Breast Cancer")
plt.show()
```

```python
# Calcular Accuracy (Precisión global)
precision = accuracy_score(test_label, predicciones)
print(f"\nAccuracy del modelo: {precision:.4f}")
```

```python
# Mostrar el Classification Report
reporte = classification_report(test_label, predicciones, target_names=data.target_names)
print("\n Reporte de Clasificación")
print(reporte)
```

```python
# Visualización del Árbol de Decisión (Resolución DPI = 300)
plt.figure(figsize=(20, 10), dpi=300)
plot_tree(
    modelo,
    feature_names=data.feature_names,
    class_names=data.target_names,
    filled=True,
    rounded=True
)
plt.title("Representación Gráfica del Árbol de Decisión")
plt.show()
```
